```python
import pandas as pd

import matplotlib.pyplot as plt


df = pd.read_csv('../DATA/Scraped_Data/medical_tests_interpretation.csv')
# Count the number of words in each row of the 'description' column
df['word_count'] = df['Description'].apply(lambda x: len(str(x).split()))

# Plot the word counts
plt.figure(figsize=(10, 6))
df['word_count'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Word Count Distribution in Descriptions')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```
